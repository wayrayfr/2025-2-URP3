# Kedzierski

```python
ked = pd.read_csv('/content/drive/MyDrive/2025/URP/data/Kedzierski et Dataset.csv')
ked = ked.drop(columns = 'Nom ')
Interpretation_counts = ked['Interpretation '].value_counts()
Interpretation_counts
```

```python
drop_list = ['Unknown']
ked = ked[~ked['Interpretation '].isin(drop_list)]
```

```python
ked['Interpretation '] = ked['Interpretation '].replace(
    {'Poly(ethylene)': 'PE',
     'Poly(ethylene) + fouling': 'PE',
     'Poly(ethylene) like': 'PE',
     'Poly(propylene)': 'PP',
     'Poly(propylene) like':'PP',
     'Poly(styrene)': 'PS',
     'Poly(amide)': 'PA',
     'Ethylene propylene rubber': 'EPR',
     'Poly(urethane)': 'PU',
     'Poly(vinylchloride)': 'PVC'})
```

```python
ked
```

```python
# Interpretation ì¹¼ëŸ¼ ì œì™¸í•˜ê³  ìˆ«ìžì¸ ì»¬ëŸ¼ë§Œ ë°˜ì˜¬ë¦¼í•´ì„œ ì´ë¦„ ë³€ê²½
new_columns = ['Interpretation'] + [str(round(float(c))) for c in ked.columns[1:]]

ked.columns = new_columns
ked
```

```python
# 1. xì¶•: ì›ëž˜ FTIR íŒŒìž¥(ì—´ ì´ë¦„)ë“¤
orig_x = ked.columns[1:].astype(float).to_numpy()   # Interpretation ì œì™¸

# 2. yì¶•: ê° í–‰ì˜ ìŠ¤íŽ™íŠ¸ëŸ¼ ê°’ (2D array: [ìƒ˜í”Œ ìˆ˜, íŒŒìž¥ ìˆ˜])
spec = ked.iloc[:, 1:].to_numpy()

# x ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬
sort_idx = np.argsort(orig_x)
orig_x_sorted = orig_x[sort_idx]
spec_sorted = spec[:, sort_idx]

new_x = np.arange(int(orig_x.min()), int(orig_x.max()) + 1)  # ì˜¤ë¦„ì°¨ìˆœ
# ë§Œì•½ 4000â†’600 ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì“°ê³  ì‹¶ìœ¼ë©´ ë§ˆì§€ë§‰ì— new_x[::-1] í•´ë„ ë¨

interp_spec = np.array([
    np.interp(new_x, orig_x_sorted, row)   # í•œ ìƒ˜í”Œ(row)ì— ëŒ€í•œ ë³´ê°„
    for row in spec_sorted
])

ked_interp = pd.DataFrame(interp_spec, columns=new_x)
ked_interp.insert(0, 'Interpretation', ked['Interpretation'].values)

```

```python
# Interpretation ì œì™¸í•œ ìˆ«ìž ì»¬ëŸ¼ë§Œ ì¶”ì¶œí•´ì„œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
sorted_cols = ['Interpretation'] + sorted(
    ked_interp.columns[1:].astype(int),
    reverse=True
)

# DataFrame ìž¬ì •ë ¬
ked_interp = ked_interp[sorted_cols]
```

```python
ked_interp
```

```python
fig = pd.read_csv('/content/drive/MyDrive/2025/URP/data/figshare.csv')
fig
```

```python
# 1) label ì»¬ëŸ¼ëª… í†µì¼
jung = jung.rename(columns={'identified': 'label'})
ked_interp = ked_interp.rename(columns={'Interpretation': 'label'})
fig = fig.rename(columns={'identified': 'label'})

# 2) ìŠ¤íŽ™íŠ¸ëŸ¼ íƒ€ìž… ì»¬ëŸ¼ ì¶”ê°€
jung['spectrum_type'] = 'transmittance'
fig['spectrum_type'] = 'transmittance'
ked_interp['spectrum_type'] = 'absorbance'

# 3) ìˆ«ìž íŒŒìž¥ ì»¬ëŸ¼ì„ ì „ë¶€ intë¡œ ê°•ì œ ë³€í™˜
def force_int_cols(df):
    rename_dict = {}
    for c in df.columns:
        if c not in ['label', 'spectrum_type']:
            rename_dict[c] = int(c)   # '4000' -> 4000, 4000.0 -> 4000
    return df.rename(columns=rename_dict)

jung = force_int_cols(jung)
ked_interp = force_int_cols(ked_interp)
fig = force_int_cols(fig)

# 4) ê° DFì—ì„œ íŒŒìž¥ ì»¬ëŸ¼ ëª©ë¡ ì¶”ì¶œ (label, spectrum_type ì œì™¸)
def get_wave_cols(df):
    return [c for c in df.columns if c not in ['label', 'spectrum_type']]

jung_waves = get_wave_cols(jung)
ked_waves = get_wave_cols(ked_interp)
fig_waves = get_wave_cols(fig)

# 5) ì „ì²´ íŒŒìž¥ ì§‘í•© ë§Œë“¤ê³  ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
all_waves = sorted(set(jung_waves) | set(ked_waves) | set(fig_waves),
                   reverse=True)

# 6) ê³µí†µ ê·¸ë¦¬ë“œë¡œ ë§žì¶”ëŠ” í•¨ìˆ˜
def align(df):
    return df.reindex(columns=['label', 'spectrum_type'] + all_waves)

jung_aligned = align(jung)
ked_aligned = align(ked_interp)
fig_aligned = align(fig)

# 7) ìµœì¢… ë³‘í•©
merged = pd.concat([jung_aligned, ked_aligned, fig_aligned],
                   axis=0, ignore_index=True)

merged
```

```python
label_counts = merged['label'].value_counts()
label_counts
```

```python
# labelë³„ count ê³„ì‚°
label_counts = merged['label'].value_counts()

# 21 ë¯¸ë§Œì¸ label ëª©ë¡
remove_labels = label_counts[label_counts < 21].index.tolist()

# í•´ë‹¹ label ì œê±°
merged_filtered = merged[~merged['label'].isin(remove_labels)].reset_index(drop=True)
```
